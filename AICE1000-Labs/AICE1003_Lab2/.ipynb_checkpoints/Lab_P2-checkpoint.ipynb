### Part 1 : Guessing Game

You are now going to work on a guessing game example. The core logic is already written however it is not easy to play.

You need to:
- Add hints for the play, for example 'too high', 'too low', and 'you're very close' when your guess is within 10 
- Keep track of how many guesses it takes to guess the number correctly, and print this number after the "Correct" message

````python
import random

random.seed(42)
number = random.randint(1, 50)

# collect the guessed number from the keyboard input
guess = int(input('Guess a number: '))

# game logic
while guess != number:
    if guess < number: 
        print('Wrong')
    if guess > number: 
        print('Wrong')
    guess = int(input('Guess a number: '))
print('Correct')
````
